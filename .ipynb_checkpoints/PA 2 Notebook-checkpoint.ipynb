# Part III: Data Preparation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error as MSE, r2_score
from sklearn import tree
```

```python
sns.set_theme(style='darkgrid')
plt.rcParams['figure.dpi'] = 200
```

```python
df = pd.read_csv('churn_clean.csv', index_col=0)
```

```python
df.info()
```

```python
with pd.option_context('display.max_rows', 6, 'display.max_columns', None): 
    display(df)
```

## Renaming columns

```python
survey_names = {'Item8': 'Active_listening', 'Item7': 'Courteous_exchange', 'Item6': 'Respectful_response', 'Item5': 'Options',
                'Item4': 'Reliability', 'Item3': 'Timely_replacements', 'Item2': 'Timely_fixes', 'Item1': 'Timely_response'}
df.rename(columns=survey_names, inplace=True)
```

```python
reformat_names = {'TimeZone': 'Time_zone', 'InternetService': 'Internet_service', 'OnlineSecurity': 'Online_security',
                 'OnlineBackup': 'Online_backup', 'DeviceProtection': 'Device_protection', 'TechSupport': 'Tech_support',
                 'StreamingTV': 'Streaming_TV', 'StreamingMovies': 'Streaming_movies', 'PaperlessBilling': 'Paperless_billing',
                 'PaymentMethod': 'Payment_method', 'MonthlyCharge': 'Monthly_charge'}
df.rename(columns=reformat_names, inplace=True)
```

## Converting 'Zip' to String and adding leading zeros

```python
df.Zip.iloc[35:40]
```

```python
df.Zip = df.Zip.astype('str')
df.Zip = df.Zip.str.zfill(5)
```

```python
df.Zip.iloc[35:40]
```

## Querying for fields containing zero-values

```python
df.loc[:, df.eq(0).any()]
```

## Checking on the 'Item' columns for input errors

```python
item_columns = df.iloc[: , -8:]
item_columns.loc[: , ~item_columns.isin(range(1,8)).any()]
```

## Converting string fields to integers

### Binary fields

```python
binary_columns = ['Churn', 'Techie', 'Port_modem', 'Tablet', 'Phone', 'Multiple',
                  'Online_security', 'Online_backup', 'Device_protection', 'Tech_support',
                  'Streaming_TV', 'Streaming_movies', 'Paperless_billing']
df[binary_columns] = df[binary_columns].replace({'Yes': 1, 'No': 0}).astype('uint8')
display(df[binary_columns])
```

## One-hot encoding

```python
df.Gender.unique()
```

```python
df.Internet_service.unique()
```

```python
df.Area.unique()
```

```python
df.Contract.unique()
```

### Gender

```python
gender_dummies = pd.get_dummies(df['Gender']).astype('uint8')
display(gender_dummies)
```

```python
df = pd.concat([df, gender_dummies], axis=1)
df.drop(['Gender', 'Nonbinary'], inplace=True, axis=1)
```

### Internet_service

```python
int_ser_dummies = pd.get_dummies(df['Internet_service']).astype('uint8')
int_ser_dummies.rename(
    columns={'Fiber Optic': 'Fiber_optic', 'None': 'No_internet'}, inplace=True)
display(int_ser_dummies)
```

```python
df = pd.concat([df, int_ser_dummies], axis=1)
df.drop(['Internet_service', 'DSL'], inplace=True, axis=1)
```

### Area

```python
area_dummies = pd.get_dummies(df['Area']).astype('uint8')
display(area_dummies)
```

```python
df = pd.concat([df, area_dummies], axis=1)
df.drop(['Area', 'Suburban'], inplace=True, axis=1)
```

### Contract

```python
contract_dummies = pd.get_dummies(df['Contract']).astype('uint8')
contract_dummies.rename(
    columns={'Month-to-month': 'Month_to_month',
             'One year': 'One_year', 'Two Year': 'Two_year'}, inplace=True)
display(contract_dummies)
```

```python
df = pd.concat([df, contract_dummies], axis=1)
df.drop(['Contract', 'One_year'], inplace=True, axis=1)
```

## Imputing missing 'Population' values

```python
df.Population = df.Population.replace(0, np.nan)
df.Population.isna().sum()
```

```python
df.Population = df.sort_values(by='Zip').Population.fillna(method='bfill')
df.Population.isna().sum()
```

## Exporting Cleaned Dataset

```python
df.to_csv('PA2_cleaned_data.csv')
```

# Part IV: Analysis

```python
X = df[['Bandwidth_GB_Year', 'Device_protection', 'Online_backup', 'Options']]
y = df.Tenure
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train.to_csv('PA2_X_train.csv')
X_test.to_csv('PA2_X_test.csv')
y_train.to_csv('PA2_y_train.csv')
y_test.to_csv('PA2_y_test.csv')
```

```python
dt = DecisionTreeRegressor(random_state=1)
```

```python
params_dt = {
    'max_depth': [5, 6, 7, 8, 9, 10],
    'min_samples_leaf': [10, 25, 50, 75, 100]
}

grid_dt = GridSearchCV(estimator=dt, param_grid=params_dt,
                       scoring='r2', cv=10, n_jobs=-1)
grid_dt.fit(X_train, y_train)
```

```python
best_model = grid_dt.best_estimator_

y_pred = best_model.predict(X_test)
test_mse = MSE(y_test, y_pred)
test_r2 = r2_score(y_test, y_pred)

print('Best model performance on test set \n')
print('Test set MSE: {:.3f}'.format(test_mse))
print('Test set R-squared: {:.3f}'.format(test_r2))
print(grid_dt.best_params_)
```

```python
min_samples = np.arange(25,65)
test_accuracy = np.empty(len(min_samples))
for i, k in enumerate(min_samples):
    test_dt = DecisionTreeRegressor(random_state=1, max_depth=8, min_samples_leaf=k)
    test_dt.fit(X_train, y_train)
    y_pred = test_dt.predict(X_test)
    test_accuracy[i] = MSE(y_test, y_pred)

plt.plot(min_samples, test_accuracy)
plt.xlabel('Min Samples per Leaf')
plt.ylabel('Test Set MSE')
plt.show()
```

```python
final_dt = DecisionTreeRegressor(random_state=1, max_depth=8, min_samples_leaf=44)
final_dt.fit(X_train, y_train)
y_pred = final_dt.predict(X_test)
test_mse = MSE(y_test, y_pred)
test_r2 = r2_score(y_test, y_pred)

print('Test set MSE: {:.3f}'.format(test_mse))
print('Test set R-squared: {:.3f}'.format(test_r2))
```

```python
plt.figure(figsize=(125,8))
tree.plot_tree(final_dt, fontsize=5, feature_names=X.columns.values)
```

### Limitation

When it comes to prediction, however, harnessing the results from multiple trees is
typically more powerful than using just a single tree. In particular, the random forest
and boosted tree algorithms almost always provide superior predictive accuracy and
performance (see “Bagging and the Random Forest” on page 259 and “Boosting” on
page 270), but the aforementioned advantages of a single tree are lost.
